```python
# FASE 5
# Calculadora de Propinas — Python Mini-Bootcamp Semana 1

print("=" * 40)
print("    CALCULADORA DE PROPINAS")
print("=" * 40)

# --- Pedir datos al usuario ---
nombre = input("¿Cómo te llamas? ")
cuenta = float(input("¿Cuál es el monto de la cuenta? $"))
porcentaje = int(input("¿Qué porcentaje de propina? (15, 18, 20): "))

# RETO A - Imprime Validación visual
print("(15) Propina estándar")
print("(18) Buena propina")
print("(20) Propina excelente")

personas = int(input("¿Entre cuántas personas se divide? "))

# --- Calcular ---
propina = cuenta * (porcentaje / 100)
total = cuenta + propina
por_persona = total / personas

# --- Mostrar resumen ---
print()
print("=" * 40)
print(f"  RESUMEN PARA {nombre.upper()}")
print("=" * 40)
print(f"  Cuenta:           ${cuenta:.2f}")
print(f"  Propina ({porcentaje}%):     ${propina:.2f}")
print(f"  Total:            ${total:.2f}")
print("-" * 40)
print(f"  Personas:         {personas}")
print(f"  Cada quien paga:  ${por_persona:.2f}")
print("=" * 40)
```

```python
# RETO B -  Múltiples monedas:
# Calculadora de Propinas — Python Mini-Bootcamp Semana 1

print("=" * 40)
print("    CALCULADORA DE PROPINAS")
print("=" * 40)

# --- Pedir datos al usuario ---
nombre = input("¿Cómo te llamas? ")
moneda = input("¿En qué moneda? (MXN/USD/EUR): ").upper()

cuenta = float(input(f"¿Cuál es el monto de la cuenta? {moneda} $"))
porcentaje = int(input("¿Qué porcentaje de propina? (15, 18, 20): "))

# RETO A - Imprime Validación visual
print("(15) Propina estándar")
print("(18) Buena propina")
print("(20) Propina excelente")

personas = int(input("¿Entre cuántas personas se divide? "))

# --- Calcular ---
propina = cuenta * (porcentaje / 100)
total = cuenta + propina
por_persona = total / personas

# --- Mostrar resumen ---
print()
print("=" * 40)
print(f"  RESUMEN PARA {nombre.upper()}")
print("=" * 40)
print(f"  Cuenta:           {moneda} ${cuenta:.2f}")
print(f"  Propina ({porcentaje}%):    {moneda}  ${propina:.2f}")
print(f"  Total:            {moneda} ${total:.2f}")
print("-" * 40)
print(f"  Personas:         {personas}")
print(f"  Cada quien paga:  {moneda} ${por_persona:.2f}")
print("=" * 40)
```
